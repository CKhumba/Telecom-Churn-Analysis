# 5.8 Measure the churn by charges

```python
#Assigning true=1 and False=0 to churn variable

df['Churn'] = df['Churn'].replace({1:bool(True),0:bool(False)})
```

```python
# Now we have to convert the categorical data into dummy variables with the getting dummies function.

telco_data_dummies = pd.get_dummies(df)
telco_data_dummies.head()
```

```python
# kdeplot to show relationship between total day charge vs the churn

Mth = sns.kdeplot(telco_data_dummies['Total day charge'][(telco_data_dummies["Churn"] == 0) ],
                color="Red", shade = True)
Mth = sns.kdeplot(telco_data_dummies['Total day charge'][(telco_data_dummies["Churn"] == 1) ],
                ax =Mth, color="Blue", shade= True)
Mth.legend(["No Churn","Churn"],loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Total day charge')
Mth.set_title('Total day charge by churn')
```

**Insight:** Here we saw that when the churn is high then the charges are high but minimal differences

```python
# kdeplot to show relationship between total eve charge vs the churn

Mth = sns.kdeplot(telco_data_dummies['Total eve charge'][(telco_data_dummies["Churn"] == 0) ],
                color="Red", shade = True)
Mth = sns.kdeplot(telco_data_dummies['Total eve charge'][(telco_data_dummies["Churn"] == 1) ],
                ax =Mth, color="Blue", shade= True)
Mth.legend(["No Churn","Churn"],loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Total eve charge')
Mth.set_title('Total eve charge by churn')
```

```python
# kdeplot to show relationship between total night charge vs the churn

Mth = sns.kdeplot(telco_data_dummies['Total night charge'][(telco_data_dummies["Churn"] == 0) ],
                color="Red", shade = True)
Mth = sns.kdeplot(telco_data_dummies['Total night charge'][(telco_data_dummies["Churn"] == 1) ],
                ax =Mth, color="Blue", shade= True)
Mth.legend(["No Churn","Churn"],loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Total night charge')
Mth.set_title('Total night charge by churn')
```

```python
# kdeplot to show relationship between Total intl charge vs the churn

Mth = sns.kdeplot(telco_data_dummies['Total intl charge'][(telco_data_dummies["Churn"] == 0) ],
                color="Red", shade = True)
Mth = sns.kdeplot(telco_data_dummies['Total intl charge'][(telco_data_dummies["Churn"] == 1) ],
                ax =Mth, color="Blue", shade= True)
Mth.legend(["No Churn","Churn"],loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Total intl charge')
Mth.set_title('Total intl charge by churn')
```
